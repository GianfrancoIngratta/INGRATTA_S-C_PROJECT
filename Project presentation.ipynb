

# The physics context

This project is actually a small part of grater ATLAS collaboration for the search of Doubly-charged Higgs boson ($H^{\pm\pm}$). In particular here the focus goes in the theoretical decay channel: $H^{\pm\pm}\rightarrow\tau^{\pm}\tau^{\pm}$ in which you have 2 isolated **same charge taus** in the final state. One is particularly interested in a dilepton signatures with the same charge like this because the majority of background at LHC is given by opposit charge dilepton signatures.



## $\tau$ decay products

If one wants to see the $H^{\pm\pm}$ decay into 2 $\tau$ is clear that, since $\tau$ decay very quickly, one has to work with the $\tau$ decay products. $\tau$ decays either in leptonic or semileptonic modes. For the reconstruction of $\tau$ decay event, it is better to focus only on semileptonic modes as lepton coming from leptonic decay are almost undistinguished from any other lepton from the background.

Within the semileptonic modes, we distinghuish:

- *1 prong decay* : $\tau$ semileptonic decay with 1 charged $\pi$ in the final state;

- *3 prong decay* : $\tau$ semileptonic decay with 3 charged $\pi$ in the final state;


## $\tau$ mis-identification probability definition

The mip is the probability, of assigning the wrong sign to decayed $\tau$ in the reconstruction analysis of the $\tau$ decay products. Knowing the mip is, of course, extremly important in the $H^{\pm\pm}\rightarrow\tau^{\pm}\tau^{\pm}$ search, because wrongly identifying the charge of one $\tau$, means classify a signal event as background and viceversa.



# Dataset

The selection criteria that define the control region, are those for which the majority of the selected event are of the type represented in the Feynman diagram above. They are:

- 2 b-jets

- 1 isolated muon

- 1 or 3 prong signatures for the tau

- energy threshold of $50$ $GeV$ (from kinematic considerations)

Then, apart from these criteria, one looks at:

- same sign (SS) for the charges of $\tau$ and $\mu$ to count the mis-reconstructed events 

- opposit sign (OS) for the charge of $\tau$ and $\mu$ to count the well-recontructed events

Once one requires these features to be satisfied, of course we still have background events (that are not $t\bar{t}$ events) which populates the control region. For the backgound composition, simulated distribution are available, except for the so called **'fakes' which is the dominant contribution for the background**.

Fakes are QCD event whose jets are wrongly reconstructed as taus. There are no reliable simulation for that, but the fakes distribution is available and given by a data-driven method using an auxilairy region ('sideband region'). For the purpose of the project we simply need to know that this is the dominant contribution for the backgound, whereas all the other contributions are treated as systematic errors.


# **mip estimate**

**The mip estimate is simply given by the number of same sign $t\bar{t}$ events over the total number of $t\bar{t}$ events.**


## Summary  dataset features for the control region

- The majority of the events are of $t\bar{t}$ type (the one we are interested to calculate the mip)

- The dominant background is given by the fakes (other bg treated as systematic errors)

- Histograms corrisponding to distributions of $t\bar{t}$ and background are available:

     * Histograms are separately given as function of two variables: transverse momentu ($p_t$) and pseudorapidity ($\eta$)
     
     * Histograms are given both for the 1 prong and 3 prong selection criteria

     * For each component which populate the control region there are 2 distributions, corrisponding to opposit-sign (OS) and same-sign(SS) different criteria that one can requires.
     
At the end of day the mip will be evalueted both for 1 and 3 prong decay and both as function of the transverse momentum and pseudorapidity.

# columns

Apart from the first column wich contains general information about the binning, each column represent a different type of event which populate the control region. The last column, ttbar, is the event of interest, the one from which the charge misidentification probability will be evaluated.

# rows (the one of interest)

## nbins: 
gives the total number of bins in the control region

## bins:
gives the bin edges in the control region

## total_fit:
gives the total number of entries, integrated all over the bins of the control region

## fixed:
stands for a variable that can be either true o false depending on whether the component is treated as systematic error or not

## curve:
gives for each component a list containing the number of entries for each bin, the namely the
 pdf 

```python
for i in range(0,len(data)):
    
    pdfs_list = data[i].loc['curve',:][1:] #first column 'metadata' is not a pdf but contains just general info
    
    bin_edges = np.array(data[i].loc['bins','metadata'])
    
    bin_centers = (bin_edges[1:]+bin_edges[:-1])/2
    
############# variable #######################################
    
    if 'pt' in data_sorted[i]:
        
        x_label='pt [GeV]'
        
    else:
        
        x_label='eta'

##############################################################

    for j in pdfs_list.index:
        
        plt.step(bin_centers[:-1],pdfs_list[j][:-1],label=str(j))
        plt.title(str(data_sorted[i]))
        plt.xlabel(x_label)
        plt.ylabel('events')
        plt.legend()
    
    plt.show()
```

As you can see the blue (ttbar) and ligth blue(fakes) components are the dominant ones, all the other pdf will be treated as systematics.
